# Modelo de regresion lineal
## Modelo con datos simulados


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = 1.5+2.5 * np.random.randn(100)
e = 0 + 0.7 * np.random.randn(100)
y_prediction = 3 + 1.8 * x
y_current = 3 + 1.8 * x * e 
```

```python
#Data Frame
data = pd.DataFrame(
        {
            "x" : x.tolist(),
            "y_current" : y_current.tolist(),
            "y_prediction" : y_prediction.tolist()
        }
)
```

```python
#obtener media actual
y_mean = [np.mean(y_current) for i in range (1,len(x))+]
```

```python
plt.plot(data["x"],data["y_prediction"])
plt.plot(data["x"],data["y_current"], "ro")
plt.plot(data["x"],y_mean,"g")
plt.title("Valor Actual vs Valor de la Prediccion")
```


# SST = Variabilidad de los datos (que tan separados de la media)

# SSD = Diferencia de datos originales  con error

# SSR = la diferencia de los dos datos

*SST = SSD + SSR

SST: Variabilidad de los datos con respecto a su media
SSD: Diferencia entre los datos originales y las predicciones que el modelo no es capaz de explicar(Errores que deberian seguir una distribucon normal)
SSR: Diferencia entre la regresion y el valor medio que el modelo busca explicar
R2 = SSR/SST coeficiente de determinacion entre 0 y 1
